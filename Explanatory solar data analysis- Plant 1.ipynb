```python
dc_gen = df_gen1.copy()
dc_gen['time'] =dc_gen['DATE_TIME'].dt.time 
dc_gen=dc_gen.groupby(['time','SOURCE_KEY'])['DC_POWER'].mean().unstack()

cmap = sns.color_palette('Spectral',n_colors=12)

fig,ax = plt.subplots(ncols=2,nrows=1,dpi=100)
fig.set_size_inches(10, 6)
dc_gen.iloc[:,0:11].plot(ax=ax[0],color=cmap)
dc_gen.iloc[:,11:22].plot(ax=ax[1],color=cmap)

ax[0].set_title('First 11 sources')
ax[0].set_ylabel('DC POWER kw')
ax[1].set_title('last 11 sources')
plt.show()
```

```python
df_gen1['time'] = df_gen1['DATE_TIME'].dt.time

fig,ax = plt.subplots(ncols=2,nrows=1,figsize=(16,9))

df_gen1.plot(x='DATE_TIME',y='DAILY_YIELD',color='green',ax=ax[0])

df_gen1.plot(x='time',y='DC_POWER',style='^',ax=ax[1])
df_gen1.plot(x='time',y='AC_POWER',style='^',ax=ax[1])
```

```python
tf_gen1=df_gen1.copy()

tf_gen1['time']=df_gen1['DATE_TIME'].dt.time
tf_gen1['day']=df_gen1['DATE_TIME'].dt.date

tf_sen1=df_sen1.copy()

tf_sen1['time']=df_sen1['DATE_TIME'].dt.time
tf_sen1['day']=df_sen1['DATE_TIME'].dt.date

#columns

col = tf_gen1.groupby(['time','day'])['DC_POWER'].mean().unstack()
print(col)
```

```python
ax = tf_gen1.groupby(['time','day'])['DC_POWER'].mean().unstack().plot(sharex=True,subplots=True,layout=(17,2),figsize=(20,30))
tf_gen1.groupby(['time','day'])['DAILY_YIELD'].mean().unstack().plot(sharex=True,subplots=True,layout=(17,2),figsize=(20,20))

i=0

for a in range(len(ax)):
    for b in range(len(ax[a])):
        ax[a,b].set_title(col.columns[i],size=15)
        ax[a,b].legend(['DC_POWER','DAILY_YIELD'])
        i=i+1

plt.tight_layout()
plt.show()
```
